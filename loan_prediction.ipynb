```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Link to download the dataset https://www.kaggle.com/datasets/adarshsng/lending-club-loan-data-csv

df = pd.read_csv("/content/loan.csv")
```

```python
df.head() #Viewing the first 5 rows of the data
```

```python
print(df.isnull().sum().to_string())
```

```python
# Calculate the threshold: 50% of the total number of rows
threshold = len(df) * 0.5

# Drop columns (axis=1) that do not meet the minimum non-null requirement
df = df.dropna(axis=1, thresh=threshold)
```

```python
print(df.isnull().sum().to_string())
```

```python
# Example of dropping leakage columns
leakage_cols = ['last_pymnt_d', 'last_pymnt_amnt', 'last_credit_pull_d', 'recoveries', 'total_rec_prncp']
df.drop(columns=leakage_cols, inplace=True)
```

```python
df.emp_title.unique().tolist()
```

```python
df.drop(columns=['application_type',], inplace=True)
```

```python
print(df.isnull().sum().to_string())
```

```python
# Create a list of columns that are known AT THE TIME of application
keep_cols = [
    'loan_amnt', 'term', 'int_rate', 'installment', 'grade', 'sub_grade',
    'emp_length', 'home_ownership', 'annual_inc', 'verification_status',
    'loan_status', 'purpose', 'addr_state', 'dti', 'delinq_2yrs',
    'earliest_cr_line', 'inq_last_6mths', 'open_acc', 'pub_rec', 
    'revol_bal', 'revol_util', 'total_acc', 'mort_acc'
]

# Filter your dataframe
df_final = df[keep_cols].copy()
```

```python
df_final.info()
```

```python
df_final.term.value_counts()
```

```python
# Strip ' months' and convert to int
df_final['term'] = df_final['term'].str.replace(' months', '').astype(int)

```

```python
df_final.grade.value_counts() 
```

```python
df_final.sub_grade.value_counts()
```

```python
# Create a sorted list of unique sub_grades
# This sorts them as A1, A2... B1, B2... G5
sub_grades = sorted(df_final['sub_grade'].unique())

# Create the dictionary mapping (A1: 1, A2: 2, etc.)
sub_grade_map = {val: i+1 for i, val in enumerate(sub_grades)}

# Apply to the dataframe
df_final['sub_grade_numeric'] = df_final['sub_grade'].map(sub_grade_map)

# Now you can drop the old string columns
df_final.drop(['grade', 'sub_grade'], axis=1, inplace=True)
```
