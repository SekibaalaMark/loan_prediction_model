```python
# 1. First, handle the '< 1 year' case specifically
# We use .replace() to turn it into a string that the regex won't confuse with '1'
df['emp_length'] = df['emp_length'].replace('< 1 year', '0 years')

# 2. Now run the extraction
# This will pull '0' for the first case and '1' for the second
df['emp_length'] = df['emp_length'].str.extract('(\d+)').astype(float)
```

```python
# 1. Define the mapping dictionary
emp_map = {
    '< 1 year': 0,
    '1 year': 1,
    '2 years': 2,
    '3 years': 3,
    '4 years': 4,
    '5 years': 5,
    '6 years': 6,
    '7 years': 7,
    '8 years': 8,
    '9 years': 9,
    '10+ years': 10
}

# 2. Apply the mapping using .loc to avoid the SettingWithCopyWarning
df.loc[:, 'emp_length'] = df['emp_length'].map(emp_map)
```

```python
# Now convert the type explicitly
df['emp_length'] = df['emp_length'].astype(float)
```

```python

# 3. Check the result - they should now be floats or ints
print(df['emp_length'].value_counts())
```

```python
# 1. Ensure they are datetime objects
df['issue_d'] = pd.to_datetime(df['issue_d'])
df['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'])

# 2. Extract the safe numerical feature
df['cr_history_months'] = ((df['issue_d'] - df['earliest_cr_line']) / np.timedelta64(1, 'M')).astype(int)

# 3. Drop the leaky/raw date columns
# We no longer need the 'dates', only the 'duration' we just created
df.drop(['issue_d', 'earliest_cr_line'], axis=1, inplace=True)
```

```python
# 1. Drop 'Current' loans AND create a deep copy
# The .copy() is the secret sauce to stop the warnings
df = df[df['loan_status'] != 'Current'].copy()

# 2. Define your binary target
def create_target(status):
    if 'Fully Paid' in status:
        return 0  # Good Loan
    else:
        return 1  # Bad Loan (Charged Off, Default, Late)

# 3. Apply the mapping using .loc (the "safe" way)
df.loc[:, 'target'] = df['loan_status'].apply(create_target)

# 4. Drop the original column
df.drop('loan_status', axis=1, inplace=True)
```
