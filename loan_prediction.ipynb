```python
# Calculate the threshold: 50% of the total number of rows
threshold = len(df) * 0.5

# Drop columns (axis=1) that do not meet the minimum non-null requirement
df = df.dropna(axis=1, thresh=threshold)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Check distribution of the strings before conversion
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='emp_length', order=['< 1 year', '1 year', '2 years', '3 years', '4 years', '5 years', '6 years', '7 years', '8 years', '9 years', '10+ years'])
plt.xticks(rotation=45)
plt.title("Distribution of Employment Length")
plt.show()
```

```python
# Use the actual mode (most frequent value)
mode_val = df['emp_length'].mode()[0]
df['emp_length'] = df['emp_length'].fillna(mode_val)
```

```python
df.emp_length.isnull().sum()
```

```python
df.emp_length.value_counts()
```

```python
# 1. First, handle the '< 1 year' case specifically
# We use .replace() to turn it into a string that the regex won't confuse with '1'
df['emp_length'] = df['emp_length'].replace('< 1 year', '0 years')

# 2. Now run the extraction
# This will pull '0' for the first case and '1' for the second
df['emp_length'] = df['emp_length'].str.extract('(\d+)').astype(float)
```

```python
# 1. Define the mapping dictionary
emp_map = {
    '< 1 year': 0,
    '1 year': 1,
    '2 years': 2,
    '3 years': 3,
    '4 years': 4,
    '5 years': 5,
    '6 years': 6,
    '7 years': 7,
    '8 years': 8,
    '9 years': 9,
    '10+ years': 10
}

# 2. Apply the mapping using .loc to avoid the SettingWithCopyWarning
df.loc[:, 'emp_length'] = df['emp_length'].map(emp_map)
```

```python
# Now convert the type explicitly
df['emp_length'] = df['emp_length'].astype(float)
```

```python

# 3. Check the result - they should now be floats or ints
print(df['emp_length'].value_counts())
```
