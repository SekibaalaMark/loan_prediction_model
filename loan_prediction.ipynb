```python
df = pd.read_csv(
    "/content/drive/MyDrive/accepted_2007_to_2018Q4.csv.gz",
    low_memory=False
)
```

```python
print(df.isnull().sum().to_string())
```

```python
df.drop(columns=['member_id','id'],inplace=True) # dropped columns that cannot be used in prediction
```

```python
print(df.isnull().sum().to_string())
```

```python
# Calculate the threshold: 50% of the total number of rows
threshold = len(df) * 0.5

# Drop columns (axis=1) that do not meet the minimum non-null requirement
df = df.dropna(axis=1, thresh=threshold)
```

```python
print(df.isnull().sum().to_string())
```

```python
df['emp_title'] = df['emp_title'].astype('category')
title_counts = df['emp_title'].value_counts()
```

```python
print(title_counts)
```

```python
df.drop(columns=['emp_title'],inplace=True)
```

```python
df.emp_length.value_counts()
```

```python
# 1. Define the mapping dictionary
emp_map = {
    '< 1 year': 0,
    '1 year': 1,
    '2 years': 2,
    '3 years': 3,
    '4 years': 4,
    '5 years': 5,
    '6 years': 6,
    '7 years': 7,
    '8 years': 8,
    '9 years': 9,
    '10+ years': 10
}

# 2. Apply the mapping using .loc to avoid the SettingWithCopyWarning
df.loc[:, 'emp_length'] = df['emp_length'].map(emp_map)
```

```python
# Now convert the type explicitly
df['emp_length'] = df['emp_length'].astype(float)
```

```python

# 3. Check the result - they should now be floats or ints
print(df['emp_length'].value_counts())
```
